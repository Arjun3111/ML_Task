```python
import pandas as pd
```

```python
dftest=pd.read_csv('cleanedtest.csv')
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder()
```

```python
dftest['Gender']=le.fit_transform(dftest['Gender'])
dftest['Married']=le.fit_transform(dftest['Married'])
dftest['Education']=le.fit_transform(dftest['Education'])
dftest['Self_Employed']=le.fit_transform(dftest['Self_Employed'])
dftest['Dependents']=le.fit_transform(dftest['Dependents'])
dftest['Property_Area']=le.fit_transform(dftest['Property_Area'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import numpy as np
```

```python
oe = OneHotEncoder(dtype=np.int, sparse=True)
```

```python
ttest = pd.DataFrame( oe.fit_transform(dftest[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']]).toarray(),
columns=['Male','Female','unmarried','Married','Independent','1Dependent','2Dependents','3+Dependents','Graduate','Ungraduate','Not_Self_Employed','Self_Employed','Rural','Semiurban',
             'Urban'])
```

```python
ttest.head()
```

```python
ttest['Loan_ID']=dftest.Loan_ID
```

```python
dftest.info()
```

```python
ttest.info()
```

```python
dffinal=pd.merge(ttest,dftest,on='Loan_ID')
```
